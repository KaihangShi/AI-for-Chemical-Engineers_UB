This demo shows the object-oriented programming in Python using class. Let's define a `Stduent` class.

```python
class Student:
    # A class representing a student

    def __init__(self, any_name, any_student_id, any_major):
        # Constructor to initialize a Student object
        self.name = any_name         # Student's name
        self.student_id = any_student_id # Student's ID
        self.major = any_major       # Student's major
        self.courses = ["Intro to ChemE"]        # A list to store courses the student is taking

    def enroll(self, course):
        # Method to enroll the student in a course
        if course not in self.courses:
            self.courses.append(course)
            print(f"{self.name} enrolled in {course}")
        else:
            print(f"{self.name} is already enrolled in {course}")

    def display_info(self):
        # Method to display student information
        print(f"Student Name: {self.name}")
        print(f"Student ID: {self.student_id}")
        print(f"Major: {self.major}")
        print(f"Enrolled Courses: {', '.join(self.courses) if self.courses else 'None'}")
```

```python
# Creating instances of the Student class
student1 = Student("Alice", "S12345", "Chemical Engineering")
```

```python
# Check native variable for each object
student1.name, student1.student_id, student1.major, student1.courses
```

```python
# Using the methods
student1.enroll("AI for Chemical Engineers")
student1.courses
```

```python
# Using the methods
student1.enroll("AI for Chemical Engineers")
```

```python
# use another method to display all information
student1.display_info()
```

Class can also inherit from a parent class. For example, the following subclass `GraduateStudent` inherit attributes and methods from parent class `Student`:

```python
class GraduateStudent(Student):
    # A class representing a graduate student, inheriting from Student

    def __init__(self, any_name, any_student_id, any_major, any_thesis_topic):
        # Constructor for GraduateStudent, calling the parent class constructor
        super().__init__(any_name, any_student_id, any_major)
        # Graduate student's unique attribute: thesis topic
        self.thesis_topic = any_thesis_topic

    def display_info(self):
        # Override display_info to include thesis topic
        # Call the parent class's display_info
        super().display_info()
        print(f"Thesis Topic: {self.thesis_topic}")
```

```python
# Creating an instance of the GraduateStudent class
grad_student1 = GraduateStudent("Bob", "G98765", "Chemical Engineering", "Advanced Simulation")
```

```python
# grad_student1 object inherits attributes from parent class `Student`
grad_student1.name, grad_student1.student_id, grad_student1.major, grad_student1.courses
```

```python
# it also inherits the methods from the parent class
grad_student1.enroll("AI for Chemical Engineers")
```

```python
# Displaying graduate student information with new thesis information
grad_student1.display_info()
```
